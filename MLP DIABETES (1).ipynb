# Multi-layer Perceptron

## Diabetes dataset


### Import dataset

```python
import pandas as pd
import numpy as np

bnotes = pd.read_csv('diabetes1.csv')
print(bnotes.head())
print(bnotes['Outcome'].unique())
```

```python
bnotes.shape
```

```python
bnotes.describe(include = 'all')
```

```python
X = bnotes.drop('Outcome', axis=1)
y = bnotes['Outcome']
print(X.head(2))
print(y.head(2))
```

### Splitting to training and testing


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3)
print(X_train.shape)
print(y_test.shape)
```

```python
#SCALING
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()  
scaler.fit(X_train)
X_train = scaler.transform(X_train)  
X_test = scaler.transform(X_test) 
print('x_train = \n',X_train)
print('x_test = \n',X_test)
```

Normalized input X train

## Train the model

Import the MLP classifier model from sklearn

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp = MLPClassifier(max_iter=100, hidden_layer_sizes=3)
mlp
```

### Training

```python
mlp.fit(X_train,y_train)
```

```python
pred = mlp.predict(X_test)
pred
```

## Evaluation metrics- Confusion matrix and F2 score

```python
from sklearn.metrics import classification_report,confusion_matrix

confusion_matrix(y_test,pred)
```

```python
print(classification_report(y_test,pred))
```
